```python
with open('./data/ko_noun_dict.pkl', "rb") as f:
    ko_dict = pickle.load(f)
    
with open('./data/en_noun_dict.pkl', "rb") as f:
    en_dict = pickle.load(f)
```

```python
def get_noun_data(ko_dict, en_dict):
    for i in range(len(en_dict)):
        ko_vector = list(ko_dict.values())[i]
        en_vector = list(en_dict.values())[i]
        
        yield (ko_vector, en_vector)
    
def get_noun_data_2(ko_vec, en_vec):
    for i in range(len(en_dict)):
        ko_vector = ko_vec[i]
        en_vector = en_vec[i]
        
        yield ko_vector, en_vector
```

```python
ko_vector = []
en_vector = []
for i in range(len(en_dict)):
    ko_vector.append(list(ko_dict.values())[i])
    en_vector.append(list(en_dict.values())[i])
```

```python
np.mean(ko_vector)
```

```python
np.mean(en_vector)
```

```python
center_ko = ko_vector - np.mean(ko_vector, 0)
center_en = en_vector - np.mean(en_vector, 0)
```

```python
norm_ko = center_ko / np.linalg.norm(center_ko)
norm_en = center_en / np.linalg.norm(center_en)
```

```python
R, s = orthogonal_procrustes(norm_ko, norm_en)
```

```python
change_ko = np.dot(norm_ko, R.T) * s
```

```python
np.sum(np.square(norm_en - change_ko))
```
